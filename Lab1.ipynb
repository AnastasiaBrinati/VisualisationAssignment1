```python
data1 = ["A","B","C","D","E"]
data2 = [3,4,7,8,9]

# NON MI FA FARE COLOR E WIDTH
fig = mpl.figure(figsize=(14,7))
# axis (subplot)
ax = fig.add_subplot()

ax.bar(data1,data2)
ax.grid()
```

```python
x_labels = ["A","B","C","D","E"]
data_y1 = [10,20,30,40,50]
data_y2 = [20,30,40,50,60]

# arange jumps
data_x = np.arange(len(x_labels))
print(data_x)
width = 0.25

# NON MI FA FARE COLOR E WIDTH
fig = mpl.figure(figsize=(14,7))
# axis (subplot)
ax = fig.add_subplot()

ax.bar(data_x - width/2 , data_y1, width, color = 'green', label = 'Data Y1')
ax.bar(data_x + width/2 , data_y2, width, color = 'orange', label = 'Data Y2')
ax.set_xticks(data_x, x_labels)
ax.legend()
ax.grid()
```

```python
# linspace can generate the values between the limits with a 
xrange = np.linspace(-10.0,10.0)
y1 = xrange**2
y2 = xrange**3

fig = mpl.figure(figsize=(9,5))
ax = fig.add_subplot()

ax.plot(xrange, y1, color="green", label = "Pow 2")
ax.plot(xrange, y2, color="orange", label = "Pow 3")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.grid()
ax.legend()
ax.set_title("Quadratic and Cubic Pow")
```

```python
data1 = [1,2,3,4,5]
data2 = [10,20,30,40,50]

fig = mpl.figure(figsize=(9,5))
ax = fig.subplots(nrows=2, ncols=2)

ax[0,0].set_title('Top Left')
ax[0,0].plot(data1)
ax[0,0].grid()
ax[0,1].set_title('Top Right')
ax[0,1].boxplot(data2)
ax[0,1].grid()
ax[1,0].set_title('Bottom Left')
ax[1,1].set_title('Bottom Right')

fig.tight_layout()
```

```python
# 3D graphing

fig = mpl.figure(figsize=(14,7))
ax = fig.add_subplot(projection='3d')

x = np.arange(-5,5,0.05)
y = np.arange(-5,5,0.05)

x,y = np.meshgrid(x,y)

z = np.cos(x)*np.sin(y)

ax.plot_wireframe(x,y,z)

# minor changes to make it kawaii

ax.plot_surface(x,y,z, cmap=mpl.cm.coolwarm, antialiased=True)
```

```python
happydf = pd.read_csv("2019.csv", sep=",")
happydf.head(10)
```

```python
happydf.shape
```

```python
x = happydf.Country_or_region
y = happydf.Score

fig = mpl.figure(figsize=(14,40))
ax = fig.add_subplot()

ax.scatter(y, x)
ax.set_title("Happiness Score")
ax.set_xlim(2,8)
ax.set_xlabel('Score')
ax.set_ylim(-1,156)
ax.grid()
```

```python
x = happydf.Score
fig = mpl.figure(figsize=(9,5))
ax = fig.add_subplot()
# analyze the distribution
ax.grid()
ax.boxplot(x)
# it doesn't tell me much, normal distribution
```

```python
x = happydf.Score
fig = mpl.figure(figsize=(9,5))
ax = fig.add_subplot()
# analyze the distribution
ax.grid()
ax.hist(x,10,rwidth=0.4, color="orange")
```

```python
# let's do a regression to find out the correlation
# simple polinomial regression of first degree (1)
# the more you overfit the harder it becomes to recreate the data from the regression

data_linreg = np.polyfit(happydf.GDP_per_capita, happydf.Score, 1)
print(data_linreg)

# y = ax + b => [a, b]
```

```python
# does having money make you happy?

x = happydf.GDP_per_capita
y = happydf.Score
fig = mpl.figure(figsize=(14,7))
ax = fig.add_subplot()

# 156: at least n points
xp = np.linspace(-0.02,1.75,156)
# polynomial evaluation : model, range/linear space
lr = np.polyval(data_linreg, xp)

ax.grid()
ax.set_xlabel("GDP")
ax.set_ylabel("Score")
ax.scatter(x,y, color="lightgreen")
ax.plot(xp, lr, color="red")
```
